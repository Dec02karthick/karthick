# Importing Libaries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
train=pd.read_csv("C:\\Users\\User\\Desktop\\tweets_train.csv")
```

```python
train.head(10)
```

```python
train.dropna(subset=['clean_text'], inplace=True)
```

```python
max_words = 10000
max_sequence_length = 100
```

```python
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(train['clean_text'])
```

```python
sequences = tokenizer.texts_to_sequences(train['clean_text'])
x=pad_sequences(sequences,maxlen=max_sequence_length)
y=train['Sentiment'].map({'Positive':0,'Neutral':1,'Negative':2})
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = Sequential()
model.add(Embedding(input_dim=max_words,output_dim=64,input_length=max_sequence_length))
model.add(SimpleRNN(units=32))
model.add(Dense(units=3,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=5,batch_size=64,validation_data=(xtest, ytest))
```
